```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score
```

```
names = ["label", "message"]
```

```
train_data = pd.read_csv(train_file_path, sep='\t', names=names)
train_data.head()
```

```
test_data = pd.read_csv(test_file_path, sep='\t', names=names)
test_data.head()
```

```
train_data['label'] = train_data['label'].map({'ham': 0, 'spam': 1})
test_data['label'] = test_data['label'].map({'ham': 0, 'spam': 1})
```

```
X_train = train_data['message']
y_train = train_data['label']
X_test = test_data['message']
y_test = test_data['label']
```

```
#Term Frequency-Inverse Document Frequency --> evaluate the importance of a word  in a document relative to a collection of documents
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```
y_train_cat = to_categorical(y_train, num_classes=2)
y_test_cat = to_categorical(y_test, num_classes=2)
```

```
#neural network
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_vec.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_vec.toarray(), y_train_cat, epochs=10, batch_size=32, validation_split=0.2)
```

```
y_pred_prob = model.predict(X_test_vec.toarray())
y_pred = y_pred_prob.argmax(axis=1)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```
def predict_message(message):
    message_vec = vectorizer.transform([message]) #transform input message into a TF-IDF vector
    pred_prob = model.predict(message_vec.toarray())[0]
    label = 'spam' if pred_prob[1] > 0.5 else 'ham'
    return [pred_prob[1], label]
```
